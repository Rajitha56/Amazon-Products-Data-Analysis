# Amazon Sales Analysis

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("amazon.csv")
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
df.info()
```

## Data Cleaning

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)
```

```python
df.info()
```

```python
df['rating'].unique()
```

```python
df['rating'].value_counts()
```

```python
df['rating'] = df['rating'].str.replace('|', '4.1').astype(float) 
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['rating_count']= df['rating_count'].str.replace(',','').astype(float)
```

```python
df['rating_count'] = df['rating_count'].fillna(df['rating_count'].median())

df['rating_count']
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['img_link', 'product_link', 'user_id', 'user_name'], inplace=True)
```

```python
df.head()
```

# Visualization

## Category Distribution

```python

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.countplot(y=df['category'], order=df['category'].value_counts().index[:10])
plt.title("Top 10 Most Common Product Categories")
plt.show()
```

## Price vs Discount Scatter Plot

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['actual_price'], y=df['discount_percentage'])
plt.title("Price vs Discount Percentage")
plt.xlabel("Actual Price")
plt.ylabel("Discount Percentage")
plt.show()
```

## Rating Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['rating'], bins=20, kde=True)
plt.title("Distribution of Product Ratings")
plt.show()
```

## Top 10 Highest-Rated Products

```python
top_rated = df.sort_values(by=['rating', 'rating_count'], ascending=[False, False]).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_rated['rating'], y=top_rated['product_name'], palette='viridis')
plt.xlabel("Rating")
plt.ylabel("Product Name")
plt.title("Top 10 Highest Rated Products")
plt.show()
```

## World Cloud of Reviews

```python
from wordcloud import WordCloud

text = " ".join(df['review_content'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Most Common Words in Reviews")
plt.show()
```
